```python
#importing libraries
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python

# Rename columns to match the Plotly Express pie chart
covid_data.rename(columns={'Date': 'date', 'Country/Region': 'country', 'Confirmed': 'cases', 'Deaths': 'deaths', 'Recovered': 'recovered'}, inplace=True)

# Aggregating the data for a specific region or time period
region_data = covid_data.groupby('country').agg({'cases':'sum', 'deaths':'sum', 'recovered':'sum'}).reset_index()

# Assuming you want to visualize the proportion for a specific country
country_name = 'Your Country Name'

# Filter data for the specific country
country_data = region_data[region_data['country'] == country_name]

# Reshape the data for Plotly Express
country_data_melted = country_data.melt(id_vars='country', value_vars=['cases', 'deaths', 'recovered'], 
                                         var_name='Category', value_name='Count')

# Create the pie chart
fig = px.pie(country_data_melted, values='Count', names='Category', 
             title=f'Proportion of COVID-19 Cases, Deaths, and Recoveries in {country_name}')
# Show the plot
fig.show()


```

```python

# Rename columns to match the Plotly Express bar chart
covid_data.rename(columns={'Date': 'date', 'Country/Region': 'country', 'Confirmed': 'cases', 'Deaths': 'deaths', 'Recovered': 'recovered'}, inplace=True)

# Aggregating the data by 'Country/Region'
country_data = covid_data.groupby('country').agg({'cases':'sum', 'deaths':'sum', 'recovered':'sum'}).reset_index()

# Create the bar chart with different colors for each metric
fig = px.bar(country_data, x='country', y=['cases', 'deaths', 'recovered'],
             title='COVID-19 Metrics by Country',
             labels={'value': 'Count', 'variable': 'Metric', 'country': 'Country'},
             color_discrete_map={'cases': 'blue', 'deaths': 'red', 'recovered': 'green'},
             barmode='group')

# Show the plot
fig.show()
```

```python

# Rename columns to match the Plotly Express line plot
covid_data.rename(columns={'Date': 'date', 'Country/Region': 'country', 'Confirmed': 'cases', 'Deaths': 'deaths', 'Recovered': 'recovered'}, inplace=True)

# Melt the dataframe to have a single 'value' column and a 'variable' column
melted_data = covid_data.melt(id_vars=['date', 'country'], value_vars=['cases', 'deaths', 'recovered'], var_name='Category', value_name='Count')

# Create the line plot
fig = px.line(melted_data, x='date', y='Count', color='Category', line_group='country',
              labels={'Count': 'Count', 'date': 'Date'},
              title='COVID-19 Cases, Deaths, and Recoveries Over Time by Country')

# Customize the layout
fig.update_layout(xaxis_title='Date', yaxis_title='Count', legend_title='Category')

# Display the plot
fig.show()
```

```python

# Rename columns to match the Plotly Express line plot
covid_data.rename(columns={'Date': 'date', 'Country/Region': 'country', 'Confirmed': 'cases', 'Deaths': 'deaths', 'Recovered': 'recovered'}, inplace=True)

# Select the latest data (optional, you may need to aggregate data if it's cumulative)
latest_data = covid_data.groupby('country').agg({'cases':'max', 'deaths':'max', 'recovered':'max'}).reset_index()

# Create the choropleth map for cases
fig_cases = px.choropleth(latest_data, locations='country', locationmode='country names', color='cases',
                          hover_name='country', color_continuous_scale='Viridis',
                          title='COVID-19 Cases by Country')

# Create the choropleth map for deaths
fig_deaths = px.choropleth(latest_data, locations='country', locationmode='country names', color='deaths',
                           hover_name='country', color_continuous_scale='Reds',
                           title='COVID-19 Deaths by Country')

# Create the choropleth map for recoveries
fig_recoveries = px.choropleth(latest_data, locations='country', locationmode='country names', color='recovered',
                                hover_name='country', color_continuous_scale='Greens',
                                title='COVID-19 Recoveries by Country')

# Show the maps
fig_cases.show()
fig_deaths.show()
fig_recoveries.show()
```
